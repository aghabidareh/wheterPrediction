```python
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error , r2_score
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
```

```python
from google.colab.drive import  mount as m
m('/content/drive/')
```

```python
data = pd.read_csv('/content/drive/My Drive/files/weatherHistory.csv')
data.head()
```

```python
data.info()
```

```python
x = data.iloc[: , 4:-1].values
y = data.iloc[: , 3].values
```

```python
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
mms = MinMaxScaler()
x = mms.fit_transform(x)
```

```python
from pickle import dump
with open('/content/drive/My Drive/models/weather/StandardScaler.h5' , 'wb') as f:
    dump(ss , f)

with open('/content/drive/My Drive/models/weather/MinMaxScaler.h5' , 'wb') as f:
    dump(mms , f)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state=1441)
```

```python
params = {
    'random_seed' : 1441,
    'iterations' : 2000,
    'verbose' : 500,
    'loss_function' : 'MAE',
    'learning_rate' : 0.1,
    'depth' : 10,
    'task_type' : 'GPU',
}
model = CatBoostRegressor(**params)
model.fit(x_train , y_train , eval_set = (x_test , y_test))
```

```python
pred = model.predict(x_test)
print(r2_score(y_test , pred)) # over fitting
print(mean_absolute_error(y_test , pred))
```

```python
model.save_model('/content/drive/My Drive/models/weather/Model.h5')
```
